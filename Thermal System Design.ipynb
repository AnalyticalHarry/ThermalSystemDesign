### 4. Microbial Growth Curve

##### 1. Bacterial Division

Only asexual reproduction is possible in bacteria and archaea, whereas both sexual and asexual reproduction is possible in eukartyotic microbes. A single cell divides into two identically sized cells in a process known as binary fission, which is most frequently used by bacteria and archaea. Multiple fission, budding, and spore production are some additional, less frequent processes that may take place.

##### 2. Growth Curve

It has been founded that bacteria will develop in a predictable manner in a closed system or batch culture (with no food added and no wastes removed), resulting in a growth curve with four distinct phases of development: the lag phase, the exponential or log phase, the stationary phase, and the death or decline phase. Furthermore, this growth curve could indeed yield generation time for a specific organism - the time it takes for the population to double.

The specifics of each growth curve (number of cells, length of each phase, rate of growth or death, overall amount of time) will differ from organism to organism or even between conditions for the same organism. However, the pattern of four distinct stages of growth will usually persist.
